## Este script sirve para generar datos de un proyectil, que se usaran en una PINN 

```python
import numpy as np
import matplotlib.pyplot as plt

def projectile_position(t, m, b, v0x, v0y, x0, y0, g=9.81):
    """
    Calcula las posiciones x(t) y y(t) de un proyectil considerando la resistencia del aire proporcional a la velocidad.

    Parmetros:
    t (float): Tiempo en el que se desea calcular la posicin.
    m (float): Masa del proyectil.
    b (float): Coeficiente de resistencia del aire.
    v0x (float): Componente x de la velocidad inicial.
    v0y (float): Componente y de la velocidad inicial.
    x0 (float): Posicin inicial en x.
    y0 (float): Posicin inicial en y.
    g (float): Aceleracin debido a la gravedad (default es 9.81 m/s^2).

    Retorna:
    (float, float): Posiciones x(t) y y(t).
    """
    
    # Calcula la posicin x(t)
    xt = (m * v0x / b) * (1 - np.exp(-b * t / m)) + x0

    # Calcula la posicin y(t)
    yt = ((m / b) * (v0y + (m * g / b)) * (1 - np.exp(-b * t / m))) - (m * g * t / b) + y0

    return xt, yt

# Parmetros iniciales
m = 1  # masa en kg
b = 0.1 # coeficiente de resistencia del aire
v0x = 12 # velocidad inicial en x en m/s
v0y = 36# velocidad inicial en y en m/s
x0 = 0.0  # posicin inicial en x en metros
y0 = 0.0  # posicin inicial en y en metros

# Tiempo total y pasos de tiempo
t_total = 6 # tiempo en segundos
t_values = np.linspace(0, t_total, 500)  # 500 pasos de tiempo entre 0 y t_total

# Listas para almacenar las posiciones
x_values = []
y_values = []

# Calcular las posiciones para cada paso de tiempo
for t in t_values:
    xt, yt = projectile_position(t, m, b, v0x, v0y, x0, y0)
    x_values.append(xt)
    y_values.append(yt)

# Graficar las posiciones
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label='Trayectoria del proyectil')
plt.title('Trayectoria del proyectil considerando resistencia del aire')
plt.xlabel('Posicin x (m)')
plt.ylabel('Posicin y (m)')
plt.legend()
plt.grid(True)
plt.show()
```
